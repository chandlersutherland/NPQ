```python
import matplotlib
matplotlib.use('Agg')
from datascience import Table
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

```python
def visualize(bleh):
    commatable = Table.read_table(bleh)
    dropped = commatable.drop(0, 1, 3)
    return dropped.plot('No.')
```

```python
def average_plot(bleh):
    NPQ_table = Table.read_table(bleh)
    avg1 = (NPQ_table.column('NPQ1') + NPQ_table.column('NPQ7'))/2
    avg2 = (NPQ_table.column('NPQ2') + NPQ_table.column('NPQ8'))/2
    avg3 = (NPQ_table.column('NPQ3') + NPQ_table.column('NPQ9'))/2
    avg4 = (NPQ_table.column('NPQ4') + NPQ_table.column('NPQ10'))/2
    avg5 = (NPQ_table.column('NPQ5') + NPQ_table.column('NPQ11'))/2
    avg6 = (NPQ_table.column('NPQ6') + NPQ_table.column('NPQ12'))/2
    avg_table = Table().with_columns('No.', NPQ_table.column('No.'), 'lhcsr 1+', avg1, 'NPQ 4 -', avg2, 'NPQ4 lhcsr 1', avg3, 'ami 1', avg4, 'ami 6', avg5, '4a 6.1 a+ WT', avg6)
    return avg_table.plot('No.')
```

```python
#3 hours HL 
plate1_control = average_plot('180831 plate 1 control.csv')
plate1_HL = average_plot('180831 plate 1 HL.csv')
print(plate1_control, plate1_HL)
```

```python
#3 days HL
plate2_control = average_plot('180831 plate 2 control.csv')
plate2_HL = average_plot('180901 plate 2 HL.csv')
print(plate2_control, plate2_HL)
```
